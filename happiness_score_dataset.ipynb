# outlier removing

# using zscore

```python
from scipy.stats import zscore
```

```python
zscore_outliers=np.abs(zscore(df))
zscore_outliers
```

```python
threshold=3
print('Outliers:- ',np.where(zscore_outliers>3))
```

```python
df_zscore=df[(zscore_outliers<3).all(axis=1)]
```

```python
print("shape of original dataframe: ",df.shape)
print("shape of dataframe using zscore: ",df_zscore.shape)
print("removal of rows: ",df.shape[0]-df_zscore.shape[0])
```

# using IQR

```python
from scipy import stats
```

```python
IQR=stats.iqr(df)
IQR
```

```python
Q1=df.quantile(0.25)
Q3=df.quantile(0.75)
IQR = df[~ ((df < (Q1 -1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
print("shape of original dataframe: ",df.shape)
print("shape of dataframe using zscore: ",IQR.shape)
print("removal of rows: ",df.shape[0]-IQR.shape[0])
```

by comparing zscore and IQR method, rows removal in IQR is less.
so we are using IQR method

```python
df=IQR
df.shape
```

# Model selection

# diving data in x and y

```python
x=df.drop(['Happiness Score'],axis=1)
x.shape
```

```python
y=df['Happiness Score']
y.shape
```

```python
# importing libraries for model selecting
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split
```

# splitting data

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.33,random_state=42)
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
lr.score(x_train,y_train)
```

```python
#predict the value
pred=lr.predict(x_test)
print("predicted happiness score : ",pred)
print("actual score : ", y_test)
```

```python
# printing errors
print("Errors:")
print("Mean absolute error : ",mean_absolute_error(y_test,pred))
print("Mean squared error : ",mean_squared_error(y_test,pred))
print("Root mean squared error : ",np.sqrt(mean_squared_error(y_test,pred)))
```

```python
print("r2 score : ",r2_score(y_test,pred))
```
