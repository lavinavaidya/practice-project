# removing outliers 

# using zscore

```python
from scipy.stats import zscore
z=np.abs(zscore(df))
z
```

```python
# threshold for zscore value, above 3 are considered ouliers
threshold=3
print(np.where(z>3))
```

```python
df_new=df[(z<3).all(axis=1)]
df_new.shape
```

```python
print('shape of df_new(without outliers): ',df_new.shape)
print('shape of df(with outliers): ',df.shape)
print('droping rows after removing outliers :',df.shape[0]-df_new.shape[0])
```

# using IQR method

```python
from scipy import stats
IQR = stats.iqr(df)
IQR
```

```python
Q1 = df.quantile(0.25)
Q3 =  df.quantile(0.75)
```

```python
df_out = df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
print(df_out.shape)
```

```python
print('shape of df_out(without outliers): ',df_out.shape)
print('shape of df(with outliers): ',df.shape)
print('droping rows after removing outliers :',df.shape[0]-df_out.shape[0])
```

from the observation of zscore and IQR method,it seen that droping of rows in IQR is high as compare to zsore method

# convert output column(quality) into categorical data

1 is classified as good wine(>=7) and 0 is classified as not good(<7)

```python
good_wine=(df['quality']>=7).value_counts()
good_wine
```

```python
notgood_wine=(df['quality']<7).value_counts()
notgood_wine
```

```python
df.loc[(df['quality'] < 7.0),'quality']=0
df.loc[(df['quality'] >=7.0),'quality']=1
```

```python
print(df.quality.value_counts())
sns.countplot(df['quality'])
```

after converting it into categorical data we observe that our data is imbalanced. so we have to make it balanced data

```python
dx=df.drop(['quality'],axis=1)
dy=df['quality']
print(dx.shape)
print(dy.shape)
```

# Hyper Perameter Tuning

```python
bgc=BaggingClassifier()
param_grid={
    'n_estimators':[100,200],
    'max_features':[1,2]
}
```

```python
bgc_grid=GridSearchCV(BaggingClassifier(),param_grid,cv=10,scoring='accuracy',n_jobs=-1,verbose=2)
```

```python
bgc_grid.fit(x_train,y_train)
bgc_pred=bgc_grid.best_estimator_.predict(x_test)
print("Accuracy after parameter tuning::",accuracy_score(y_test,bgc_pred))
```
